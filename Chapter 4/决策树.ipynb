```python
from sklearn import tree
from sklearn.datasets import load_iris
import graphviz
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
iris = load_iris()
data, target = iris.data, iris.target
print(data.shape, target.shape)
print(set(target))
```

```python
train_data, test_data, train_label, test_label = train_test_split(data, target, test_size=0.3, random_state=1)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_label)
train_predict = clf.predict(train_data)
print("train accuracy:{0:.2f}".format(np.mean((train_predict==train_label).astype(np.float32))))
```

```python
test_predict = clf.predict(test_data)
print("test accuracy:{0:.2f}".format(np.mean((test_predict==test_label).astype(np.float32))))
```

```python
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph
```
